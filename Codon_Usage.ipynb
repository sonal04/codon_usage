```python
synonymous_codon= CodonUsage.SynonymousCodons.copy()
def codon_usage(c):
    for aa in synonymous_codon:
        codons = synonymous_codon[aa]
        for codon in codons:
            if c == codon:
                
                return (aa, codon,rscu[codon])
```

```python
pairs = {'A':'T','C':'G','G':'C','U':'A'}
def codon(anticodon):
    """returns the codon reverse complement for a given anticodon"""
    complement = ''.join(pairs[c] for c in anticodon)
    return complement[::-1]

codon('UGC')
```

### Anticodon freq and codon usage

```python
ac={}
ac1=[]
import pandas as pd
for i in range(0, len(test_bitome.genes)):
    if test_bitome.genes[i].gene_type == 'tRNA':
        #print (test_bitome.genes[i].trna_anticodon, test_bitome.genes[i].trna_name, test_bitome.genes[i].name)
        if test_bitome.genes[i].trna_anticodon == "NA1":
                test_bitome.genes[i].trna_anticodon = "CGG"
        if test_bitome.genes[i].trna_anticodon is not None:
            
            ac1.append(test_bitome.genes[i].trna_anticodon)
            ac.setdefault(test_bitome.genes[i].trna_name, [])
            ac[test_bitome.genes[i].trna_name].append(test_bitome.genes[i].trna_anticodon)
            #print (ac)
        
import itertools
# from collections import Counter
# for elements in ac:
#print ("Anticodon", "freq", "Codon", "Codon Usage")  
res =[]
for (key,group) in itertools.groupby(sorted(ac1)): 
    #print (codon_usage(codon(key)))
    #print (key)
    anticodon_freq= key,str(len(list(group))), codon(key), codon_usage(codon(key))
                                                                      
                                                                    
    res.append(anticodon_freq)
    #print (", ".join(res))
    #df = pd.DataFrame(res)

df = pd.DataFrame(res, columns=["Anticodon", "freq", "Codon", "Codon Usage"])
df.sort_values('Codon Usage')
```
